```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(0)
```

```python
unif = np.random.uniform(0, 10, 1000)
plt.hist(unif, bins=100)
plt.show()
```

```python
# 연속확률 분포
normal_ = np.random.normal(0, 1, 1000)
plt.hist(normal_, bins=50)
plt.show()
```

```python
expo = np.random.exponential(10, 100)
```

```python
plt.hist(expo, bins=10)
plt.show()
```
