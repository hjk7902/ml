### 상관분석

```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris.corr(method="pearson", numeric_only=True) # spearman 도 있음
```

```python
# -1 ~ 1 사이의 값 출력
# 0~0.3 약한 양의 상관관계
# 0.3~0.6(0.7) 강한 양의 상관관계
# 0.6(0.7) ~ 1 매우 강한 양의 상관관계
```

```python
sns.pairplot(iris)
```

```python
from scipy.stats import pearsonr
pearsonr(iris.petal_length, iris.petal_width)
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width")
```

```python
iris.corr(method="spearman", numeric_only=True) # 스피어만 상관계수, 정규분포를 벗어난 경우 사용
```

```python
from scipy.stats import spearmanr
spearmanr(iris.petal_length, iris.petal_width)
```
